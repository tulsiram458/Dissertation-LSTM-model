1. Importing essential libraries for preprocessing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading Britrish petroleum data from the csv file to dataframe

df = pd.read_csv(r"C:\Users\yunus\Desktop\tulasi\BP.CSV")
```

```python
# top 5 data of the dataframe df, which contains british petroleum data

df.head()
```

```python
# last 5 data of the dataframe df, which contains british petroleum data

df.tail()
```

```python
# feature selection
# copying the essential values of the dataframe into a new dataframe, as only these two features are required for processing

df2= df[['Date', 'Adj Close']].copy()
```

```python
df2.head()
```

```python
# renaming the column name from 'Adj Close' to 'BP'

df2.rename({'Adj Close' : 'BP'}, axis = 1, inplace =True)
```

```python
df2.head()
```

```python
# knowing the data types of dataframe

df2.dtypes
```

```python
# shape of the data (which will know, how much data we have in the dataframe like columns ans rows)

df2.shape
```

```python
df2.dtypes
```

```python
# changing the data structure and date format from object to datetime

df2['Date'] = pd.to_datetime(df2['Date'], format='%m/%d/%Y')
```

```python
df2.head()
```

```python
# changing the index to date as the values will be assigned or indexed based upon the Date not on numbers
# this is essential when we are doing the time series processing
# the indexed from Date has been saved in the new dataframe df3.

df3 = df2.sort_values("Date")
df3 = df2.set_index("Date")
df3.head()
```

```python
# plotting the dataframe for visualization using matplotlib 

plt.style.use('seaborn-dark')

df3.plot(cmap= "viridis",figsize=(20,15))
plt.grid()
plt.show()
```
